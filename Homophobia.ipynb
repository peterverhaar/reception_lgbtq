```python
import json
import re
from nltk import sent_tokenize,word_tokenize
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer 
ana = SentimentIntensityAnalyzer()
from tqdm import tqdm

lgbtq_lexicon = []
with open('lgbtq_lexicon.txt',encoding='utf-8') as fh:
    for term in fh:
        lgbtq_lexicon.append(term.strip()) 
```

```python
# Positive reviews, reference to LGBTQ, negative

corpus = []
json_data = json.load(open('reviews.json'))
for review in tqdm(json_data):
    review['text'] = re.sub(r'\.','. ',review['text'])
    sentences = sent_tokenize(review['text'])
    for sentence in sentences:
        

        lgbtq = False
        for term in lgbtq_lexicon:
            if re.search(r'\b{}\b'.format(term),sentence,re.IGNORECASE):
                lgbtq = True

        if lgbtq:
            scores = ana.polarity_scores(sentence)
            if scores['compound'] < -0.7:
                corpus.append(sentence)
                
```

```python
len(corpus)
```

```python
with open('lgbtq_corpus.txt','w',encoding='utf-8') as out:
    for sentence in corpus:
        out.write(f'{sentence}\n')
```
